```python
from PIL import Image
from PIL import ImageEnhance
```

```python
img = Image.open("logo.png")
img, img.size
```

### 이미지 밝기 조정

```python
def change_brightness(img, factor):
  bright_enhancer = ImageEnhance.Brightness(img)
  img_bright = bright_enhancer.enhance(factor)
  
  return img_bright
  
change_brightness(img, factor = 1.5)
```

```python
change_brightness(img, factor = 0.5)
```

### 이미지 대조 조정

```python
def change_contrast(img, factor):
  contrast_enhancer = ImageEnhance.Contrast(img)
  img_contrast = contrast_enhancer.enhance(factor)
  
  return img_contrast

change_contrast(img, 3.)
```

```python
change_contrast(img, 0.3)
```

### 이미지 크롭핑(Cropping)

```python
def crop(img, coordinates):
  img_crop = img.crop(coordinates)
  
  return img_crop 

# 이미지 자를 때 (left, upper, right, lower) 네 개의 값을 포함하는 튜플 전달 
# 만약 (left = 50, right = 200)이라면, right - left = 150만큼 원본 이미지를 잘라냄.
crop(img, coordinates = (50, 70, 200, 350))
```

```python
crop(img, coordinates=(0,0,300,300))
```

### 이미지 회전

```python
def rotate(img, angle, expand = False):
  img_rotate = img.rotate(angle, expand = expand)
  
  return img_rotate

# angle = 160 -> 시계 반대 방향으로 160도 회전
# expand: 
rotate(img, angle = 160, expand = True)
```

```python
rotate(img, angle = 180, expand = True)
```

### 이미지 크기 변경

```python
def resize(img, new_size):
    img_resize = img.resize(new_size)
    
    return img_resize
  
resize(img, (200, 200))
```

```python
resize(img, (500,500))
```
