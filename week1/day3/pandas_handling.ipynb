## 그룹화 (Groupby)
- DataFrame에서 `groupby()`를 호출해 반환된 결과에 aggregation함수를 호출하면 `groupby()`대상 컬럼을 제외한 모든 컬럼에 해당 aggregation 함수를 
- SQL과는 다르게 여러 aggregation 함수를 적용하는 경우, `agg()`를 사용해야함 

```python
df_groupby = df.groupby(by = "Pclass").count()
df_groupby
```

```python
# Pclass로 그룹화된 객체에 PassengerId와 Survived만 필터링하여 count 계산
df_group_by = df.groupby("Pclass")[["PassengerId","Survived"]].count()
df_group_by
```

```python
## SQL과는 다르게 여러 aggregation 함수를 적용하는 경우, agg()를 사용해야함 
df.groupby('Pclass')['Age'].agg([max, min])
```

```python
## 각 컬럼들에 대해 서로 다른 aggregation 함수 적용 가능
# 딕셔너리 형태로 입력하여 필터링 가능 
agg_format = {'Age':'max', "SibSp":'sum', "Fare":"mean"}
df.groupby("Pclass").agg(agg_format)
```

## DataFrame 결합

- merge: 기본적으로 내부조인
    - on -> 특정 변수를 key로 지정하여 결합 가능
    - how -> inner, outer로 내부, 외부조인 가능
    - left.merge(right)
    - left.merge(right, how = 'inner')
    - left.merge(right, on = 'key')
- join: 기본적으로 왼쪽조인
    - left.join(right)
    - left.join(right, how = 'outer')
- concat: 기본적으로 행을 기준으로 결합
    - merge, join과 다르게 pd뒤에 붙음
    - pd.concat([left, right])
    - pd.concat([left,right], axis = 1)

```python
## left, right 데이터프레임 생성

left_data = pd.DataFrame({'key':['a','b','c','d','e'],
                         'var_1':[1,3,4,2,1],
                         'var_2':[1,2,4,3,0]})


right_data = pd.DataFrame({'key':['b','c','e','f','g'],
                          'var_3':[4,6,3,2,3],
                          'var_4':[3,5,8,7,4]})
```

```python
# 내부조인으로 공통된 key의 값만 merge
res1 = left_data.merge(right_data)
res2 = left_data.merge(right_data, how = 'inner')

print(f"res1:\n {res1}\n")
print(f"res2:\n {res2}")
```

```python
# 외부조인으로, 모든 키 a ~ g값을 조인
# 동일한 key가 없는 경우 NaN 값으로 채워진다.
# 외부조인 정의상 릴레이션은 속성의 집합과 튜플의 집합이므로 key는 중복된 값을 가질 수 없다.
left_data.merge(right_data, how = 'outer')
```

```python
# 왼쪽 조인으로, 왼쪽 key값만 결합
# 오른쪽에 없는 왼쪽 key의 값은 모두 NaN 값으로 처리됨
left_data.merge(right_data, how = 'left')
```

```python
# 오른쪽 조인으로, 오른쪽 key 값만 결합
left_data.merge(right_data, how = 'right')
```

```python
## join
## error는 join에서 발생하는데, join은 인덱스를 기준으로 합치므로 key 컬럼은 key 역할이 아닌 그저 변수의 역할이다
## 따라서 이름이 겹치게되므로 합칠 수 없다는 에러가 발생함

left_data.join(right_data)
```

```python
## key 역할을 해주기 위해서 인덱스로 설정을 해줘야됨

left_data = left_data.set_index('key')
right_data = right_data.set_index('key')

left_data.join(right_data) # join은 기본적으로 왼쪽 조인
```

```python
res1 = pd.concat([left_data, right_data], axis = 0, join = "inner") # axis = 0이므로 밑에 열에 추가
print(f"res1:\n {res1}") 
```

```python
res2 = pd.concat([left_data, right_data], axis = 1, join = 'inner') # axis = 1이므로 옆에 결합
print(res2)
```
