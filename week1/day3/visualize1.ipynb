```python
import matplotlib.pyplot as plt

# matplotlib은 기본적으로 한글 지원 x
# 나눔고딕 
plt.rcParams["axes.unicode_minus"] = False
plt.rc("font", family="NanumGothicCoding")

plt.plot([1,3,5],[2,4,6])
plt.title("Hello matplotlib")

plt.show()
```

## pyplot에서 Figure와 Axes

- Figure: 그림을 그리기 위한 캔버스 역할을 하는 객체
- Axes: 실제 그림을 그리는 메서드로, 대부분의 시각화를 수행하는 메서드들은 Axes에서 호출됨(X축, Y축, 타이틀, 범례 등)
  - `plt.plot()`의 경우 `Axes.plot()`함수를 호출하여 그림을 그림
  - `plt.title()`의 경우 내부적으로 `Axes.set_title()`함수를 호출하여 타이틀 설정

```python
plt.figure(figsize = (10,6))
plt.plot([1,3,5],[2,4,6])
plt.title("Hello plot")

plt.show()
```

```python
# 여기서도 그림판 전체가 노란색으로 변하진 않음. Axes가 차지하는 영역이 Figure가차지하는 그림판 위에 적용되므로.
plt.figure(figsize = (10,6), facecolor = 'yellow')
plt.plot([1,3,5],[2,4,6])
plt.title("안녕 plot")
plt.show()
```

## 여러 개의 plot을 가지는 subplot 생성

- `subplots()`: 인자로 여러 개의 `Axes`를 설정할 수 있음, 즉 하나의 Figure상에서 여러 개의 그래프를 그릴 수 있게 만들어주는데, 이때 개별 subplot은 하나의 Axes를 가지게 된다.
  - `nrows`: 전체 subplot들의 배치를 2차원 행렬 형태로 표현할 때 행의 개수
  - `ncols`: 열의 개수

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (6,3)) # 전체 크기 가로 6, 세로 3
```

```python
import numpy as np
import pandas as pd

x_value = [1,2,3,4] # ndarray로도 사용 가능 
y_value = [2,4,6,8]

plt.plot(x_value, y_value, color = 'green') # plot 색상 변경 가능 
plt.show()
```

## plot의 다양한 인자
- `color`: plot의 색상 변경
- `marker`: x값에 따른 y값의 위치에 존재하는 마커로, 형태 정의 가능
- `linestyle`: 실선, 점선 등으로 정의 가능
- `linewidth`: 선의 두께
- `markersize`: 위에서 정의한 marker의 크기 정의 

```python
plt.plot(
  x_value, y_value, color = 'red', marker='o', linestyle= 'dashed', linewidth = 2, markersize = 12)
```

## 축 명칭, 범례 등 설정

```python
plt.plot(
  x_value, y_value, color = 'green', marker = 'o', linestyle = 'dashed', linewidth = 2, markersize = 10
)

# x축, y축에 이름 명시, fontsize로 크기 설정 가능 
plt.xlabel('x axis', fontsize = 15)
plt.ylabel('y axis', fontsize = 15)

# 만약 x축의 문자열이 긴 경우, 보기 안 좋을 수 있으므로 x축 문자의 회전이 필요할 수 있다.

plt.xticks(rotation = 90)
```

```python
# matplotlib은 min/max 값에 기반하여 축의 눈금 값들을 표시하므로, 세밀하게 나타내기 위해 ticks 인자 추가

x_values = np.arange(0,100)
y_values = 2 * x_values

plt.plot(x_values, y_values, color = 'green')
plt.xlabel("x axis")
plt.ylabel('y axis')
# X축의 눈금값을 0부터 99까지 10스텝으로 10개를 표시
plt.xticks(ticks = np.arange(0,100,10), rotation = 90)
plt.title('Hello plot')

plt.show()
```

```python
x_value_01 = np.arange(0,100)
y_value_01 = 2 * x_value_01
y_value_02 = 4 * x_value_01

plt.plot(x_value_01, y_value_01, color = 'green')
plt.plot(x_value_01, y_value_02, color = 'red')

plt.legend(["A","B"]) # 범례 지정 가능 
plt.show()
```

```python
# legend 에 정의하지 않고 plot 내부에서 label 인자에 전달할 수 있음 
plt.plot(x_value_01, y_value_01, color = 'green', label = "A")
plt.plot(x_value_01, y_value_02, color = 'red', label = "B")

plt.legend()
plt.show()
```

```python
# 직접 Axes ㄱ개체 메서드를 사용하여 시각화 수행 

x_value_01 = np.arange(1,10)
x_value_02 = np.arange(1,20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (6,6))

ax[0][0].plot(x_value_01, y_value_01, color = 'green', marker = 'o', linestyle = 'dashed', label = 'green line')
ax[0][1].bar(x_value_02, y_value_02, color = 'red', label = "red bar")
ax[1][0].plot(
  x_value_01, y_value_01, color = 'red', marker = 'o', linestyle = 'dashed', label = 'red line'
)
ax[1][1].bar(x_value_02, y_value_02, color = 'green', label = 'green bar')

ax[0][0].set_xlabel('ax[0][0] x axis')
ax[0][1].set_xlabel('ax[0][1] x axis')
ax[1][0].set_xlabel('ax[1][0] x axis')
ax[1][1].set_xlabel('ax[1][1] x axis')

ax[0][0].legend()
ax[0][1].legend()
ax[1][0].legend()
ax[1][1].legend()

plt.show()
```

```python
fig1 = plt.figure(1, figsize = (4,3))
plt.plot(x_value_02, y_value_02, label = "A")
plt.title("fig1")

fig2 = plt.figure(2, figsize = (4,3))
plt.plot(x_value_01, y_value_01)
plt.title("fig")

plt.show()
```

## Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["axes.unicode_minus"] = False
plt.rc("font", family="NanumBarunGothic")
```

```python
# 데이터 리스트로 생성
year = [2018, 2019, 2020, 2021, 2022, 2023]
python = [5.8, 8.17, 9.31, 11.87, 15.63, 14.51]
C = [13.59, 14.08, 16.72, 14.32, 12.71, 14.73]
java = [15.78, 15.04, 16.73, 11.23, 10.82, 13.23]
JS = [3.49, 2.51, 2.38, 2.44, 2.12, 2.1]
```

```python
sns.lineplot(x = year, y = python)
```

seaborn은 matplotlib를 기반으로 개발된 라이브러리이기 때문에 matplotlib의 매개변수 대부분을 사용할 수 있음

```python
sns.lineplot(x = year, y = python, label = 'python')
sns.lineplot(x = year, y = java, label = 'java')
```

`sns.lineplot()` 함수는 `matplotlib.axes` 를 반환하는 함수로,
`matplotlib.axes`의 매소드들을 이용해 그래프의 모양을 조절할 수 있음

```python
sns.set_style("darkgrid") # 테마 설정 가능

lp = sns.lineplot(x = year, y = python, label = 'python')

lp.set_xlabel("연도")
lp.set_ylabel("점유율")
```

```python
# constrained_layout = True로 지정하면 크기 비율 최적화 
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True)  # 1행 2열의 서브플롯을 생성합니다.

sns.lineplot(x=year, y=python, label="python", ax=ax1)  # seaborn을 이용해 그래프 그리기
sns.lineplot(x=year, y=java, label="java", ax=ax2)  
```

```python
import seaborn as sns
import pandas as pd

df = pd.read_csv("../../data/train.csv")
df.head()
```

```python
## histogram
import matplotlib.pyplot as plt

plt.figure(figsize = (10,6)) # seaborn에서도 figure로 크기 정의

# df의 컬럼명을 자동으로 인식해서 xlabel 값 할당 
# y axis는 빈도(count)
sns.histplot(df['Age'], bins = 20) # bins는 20개로 나누는 것
plt.show()
```

```python
# kde: 연속 확률분포 곡선 그리기 
sns.histplot(df['Age'], bins = 30, kde = True)
```

## `sns.displot()`
seaborn의 또 다른 히스토그램 함수로, `displot()`함수는 Figure 레벨 함수이다.
- matplotlib API 사용 최소화
- Figure 레벨 함수 사용 시 Figure의 크기를 `plt.figure()`로 조정할 수 없음

```python
## displot

sns.displot(df['Age'], kde = True, height = 4, aspect = 2)
```
